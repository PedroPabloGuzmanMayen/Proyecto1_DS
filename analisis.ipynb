## Análisis inicial sobre datos del MINEDUC

```python
import pandas as pd
```

```python
df = pd.read_csv("establecimientos.csv")

df.head()

```

```python
df.shape

print(f"El número de filas es de: {df.shape[0]} ")
print(f"El número de columnas es de: {df.shape[1]}")
```

```python


def resumen_categoricas_con_na(df):
    resumen = []

    for col in df.columns:
        datos = df[col]
        total = len(datos)
        no_nulos = datos.notna().sum()
        nulos = datos.isna().sum()
        n_unicos = datos.nunique(dropna=True)

        # Obtener el valor más frecuente 
        valor_frecuente = datos.mode(dropna=False)
        valor_frecuente = valor_frecuente.iloc[0] if not valor_frecuente.empty else None

        # Contar frecuencia de ese valor 
        frecuencia = datos.value_counts(dropna=False).iloc[0] if not datos.empty else 0

        # Mostrar NAN
        if pd.isna(valor_frecuente):
            valor_frecuente_str = "NaN"
        else:
            valor_frecuente_str = str(valor_frecuente)

        resumen.append({
            "Variable": col,
            "Total filas": total,
            "No Nulos": no_nulos,
            "Nulos": nulos,
            "Valores únicos (no nulos)": n_unicos,
            "Más frecuente (conteo)": f"{valor_frecuente_str} ({frecuencia} veces)"
        })

    return pd.DataFrame(resumen)

df_resumen = resumen_categoricas_con_na(df)
print(df_resumen)
```
