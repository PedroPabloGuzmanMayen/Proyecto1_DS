## Análisis inicial sobre datos del MINEDUC

```python
import pandas as pd
```

```python
df = pd.read_csv("establecimientos.csv")

df.head()

```

```python
df.shape

print(f"El número de filas es de: {df.shape[0]} ")
print(f"El número de columnas es de: {df.shape[1]}")
```

```python
resumen = []

for col in df.columns:
    datos = df[col].dropna()  # Ignorar NAs

    if datos.empty:
        resumen.append({
            "Variable": col,
            "No Nulos": 0,
            "Valores únicos": 0,
            "Valor más frecuente": None,
            "Frecuencia": 0
        })
    else:
        valor_frecuente = datos.mode().iloc[0]
        frecuencia = datos.value_counts().iloc[0]

        resumen.append({
            "Variable": col,
            "No Nulos": datos.count(),
            "Valores únicos": datos.nunique(),
            "Valor más frecuente": valor_frecuente,
            "Frecuencia": frecuencia
        })

resumen

```
